```python
from langchain.embeddings import HuggingFaceEmbeddings

# Option 1: Using a Sentence Transformers model (recommended)
model_name = "sentence-transformers/all-mpnet-base-v2"  # You can choose others like 'all-mpnet-base-v2'

hf_embed_model = HuggingFaceEmbeddings(model_name=model_name)
```

```python
import gzip
import json
wikipedia_filepath='/content/simplewiki-2020-11-01.jsonl'
docs=[]
with open(wikipedia_filepath,'rt',encoding='utf8') as f:
  for line in f:
    data=json.loads(line.strip())

    docs.append({
        'metadata':{
            'title':data.get('title'),
            'article_id':data.get('id')
        },
        'data': ' '.join(data.get('paragraphs')[0:3])
    })
print('passages:',len(docs))
```

```python
docs=[doc for doc in docs for x in ['linguistics','india','cheetah']if x in doc['data'].lower().split()]
```

```python
len(docs)
```

```python
docs[:3]
```

```python
from langchain.docstore.document import Document
docs=[Document(page_content=doc['data'],metadata=doc['metadata'])for doc in docs]
```

```python
docs[0:3]
```

```python
len(docs)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=100)
chunked_docs=splitter.split_documents(docs)
```

```python
chunked_docs[0:3]
```

```python
len(chunked_docs)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

# Recommended model similar in quality to OpenAI's ada-002 for many use cases
model_name = "sentence-transformers/all-MiniLM-L6-v2"

# Create the embedding model
openai_embed_model = HuggingFaceEmbeddings(model_name=model_name)
```

```python
from langchain_chroma import Chroma
chroma_db=Chroma.from_documents(documents=chunked_docs,collection_name='wikipidia_db',embedding=openai_embed_model,collection_metadata={"hnsw:space":"cosine"},persist_directory='./wikipedia_db')
```

```python
# load from disk
chroma_db = Chroma(persist_directory="./wikipedia_db",
                   collection_name='wikipedia_db',
                   embedding_function=openai_embed_model)
```

```python
chroma_db
```

```python
similarity_retriever = chroma_db.as_retriever(search_type="similarity",
                                              search_kwargs={"k": 3})
```

```python
similarity_retriever
```

```python
query = "what is the pink city of india"
top3_docs = similarity_retriever.invoke(query)
top3_docs
```

```python
query = "what is the old capital of India?"
top3_docs = similarity_retriever.invoke(query)
top3_docs
```
